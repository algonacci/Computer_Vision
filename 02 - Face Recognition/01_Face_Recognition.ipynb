### Import Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Face Recognition

```python
img = cv2.imread('male.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
haar = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
face = haar.detectMultiScale(gray ,1.3, 5)
print(face)
```

```python
# cv2.rectangle(img,(154,94),(154+261,94+261),(0,255,0),5)
for x,y,w,h in face:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
plt.imshow(img)
```

```python
cv2.imshow('object_detect',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
face_crop = img[94:94+261,154:154+261]
plt.imshow(face_crop)
```
