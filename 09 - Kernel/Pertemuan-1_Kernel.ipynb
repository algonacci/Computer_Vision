### Creating a Filter and make Edge Detection

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline
```

```python
plt.rcParams.update({
    "lines.color": "white",
    "patch.edgecolor": "white",
    "text.color": "white",
    "axes.facecolor": "black",
    "axes.edgecolor": "white",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "grid.color": "grey",
    "figure.facecolor": "black",
    "figure.edgecolor": "black",
    "savefig.facecolor": "black",
    "savefig.edgecolor": "black"})
```

```python
curved_lane = mpimg.imread('curved_lane.jpeg')
bridge_trees = mpimg.imread('bridge_trees_example.jpeg')

print('Curved lane shape:', curved_lane.shape)
print('Bridge trees shape:', bridge_trees.shape)
```

```python
plt.figure(figsize=(15,15))

plt.subplot(1,2,1)
plt.title('Curved lane', fontsize=20, pad=20)
plt.imshow(curved_lane)

plt.subplot(1,2,2)
plt.title('Bridge trees', fontsize=20, pad=20)
plt.imshow(bridge_trees)

plt.show()
```

```python
# Convert grayscale
plt.figure(figsize=(15,15))

curved_lane_gray = cv2.cvtColor(curved_lane, cv2.COLOR_RGB2GRAY)
plt.subplot(2,2,1)
plt.title('Curved lane', fontsize=20, pad=20)
plt.imshow(curved_lane_gray)

bridge_trees_gray = cv2.cvtColor(bridge_trees, cv2.COLOR_RGB2GRAY)
plt.subplot(2,2,2)
plt.title('Bridge trees', fontsize=20, pad=20)
plt.imshow(bridge_trees_gray)

curved_lane_gray = cv2.cvtColor(curved_lane, cv2.COLOR_RGB2GRAY)
plt.subplot(2,2,3)
plt.title('Curved lane grayscaled', fontsize=20, pad=20)
plt.imshow(curved_lane_gray, cmap='gray')

bridge_trees_gray = cv2.cvtColor(bridge_trees, cv2.COLOR_RGB2GRAY)
plt.subplot(2,2,4)
plt.title('Bridge trees grayscaled', fontsize=20, pad=20)
plt.imshow(bridge_trees_gray, cmap='gray')

plt.show()
```

```python
print('Curved lane shape:', curved_lane_gray.shape)
print('Bridge trees shape:', bridge_trees_gray.shape)
```

### Use Sobel Kernel to Extracting Edge Features

```python
sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])

sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

sobel_xy = sobel_x * sobel_y
```

```python
plt.figure(figsize=(15,15))

curved_lane_sobel_x = cv2.filter2D(curved_lane_gray, -1, sobel_x)
plt.subplot(3,2,1)
plt.title('Sobel X - Curved Lane', fontsize=20, pad=20)
plt.imshow(curved_lane_sobel_x, cmap='gray')

bridge_trees_sobel_x = cv2.filter2D(bridge_trees_gray, -1, sobel_x)
plt.subplot(3,2,2)
plt.title('Sobel X - Bridge Trees', fontsize=20, pad=20)
plt.imshow(bridge_trees_sobel_x, cmap='gray')

curved_lane_sobel_y = cv2.filter2D(curved_lane_gray, -1, sobel_y)
plt.subplot(3,2,3)
plt.title('Sobel Y - Curved Lane', fontsize=20, pad=20)
plt.imshow(curved_lane_sobel_y, cmap='gray')

bridge_trees_sobel_y = cv2.filter2D(bridge_trees_gray, -1, sobel_y)
plt.subplot(3,2,4)
plt.title('Sobel Y - Bridge Trees', fontsize=20, pad=20)
plt.imshow(bridge_trees_sobel_y, cmap='gray')

curved_lane_sobel_xy = cv2.filter2D(curved_lane_gray, -1, sobel_xy)
plt.subplot(3,2,5)
plt.title('Sobel XY - Curved Lane', fontsize=20, pad=20)
plt.imshow(curved_lane_sobel_xy, cmap='gray')

bridge_trees_sobel_xy = cv2.filter2D(bridge_trees_gray, -1, sobel_xy)
plt.subplot(3,2,6)
plt.title('Sobel XY - Bridge Trees', fontsize=20, pad=20)
plt.imshow(bridge_trees_sobel_xy, cmap='gray')

plt.show()
```
