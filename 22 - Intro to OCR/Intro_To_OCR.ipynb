```python
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Read in images or videdo

```python
image = 'algonacci.jpg'
```

```python
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(image)
result
```

### Draw Results

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
result_image = cv2.imread(image)
result_image = cv2.rectangle(result_image, top_left, bottom_right, (0,255,0), 5)
result_image = cv2.putText(result_image, text, top_left, font, .5, (255,0,0), 2, cv2.LINE_AA)
plt.figure(figsize=(10,10))
plt.imshow(result_image)
plt.show()
```

### Handling Multiple Lines

```python
img = cv2.imread('test.png')
spacer = 300
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(img)
result
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(120,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    print(text)
    spacer+=15
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```
